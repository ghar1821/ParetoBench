# Interpreting the result

ParetoBench is written in Python, but the CDF and swarm plots used in the manuscript is generated using ggplot, just because it's prettier. 

The function to draw them is stored in the plots folder of ParetoBench.

The following tutorials assumed you have run ParetoBench and have stored all the csv files somewhere. 
If not, please follow the other python notebook tutorial and generate some data.

## Sourcing libraries

Assuming the scripts are stored in ~/Documents/phd/code/ParetoBench/ParetoBench/plots, let's source the files.

***NOTE*** if you don't have ggplot2 and ggbeeswarm installed, please install them by running install.packages(c("ggplot2", "ggbeeswarm")).


```r
library(ggplot2)
library(ggbeeswarm)

setwd("~/Documents/phd/code/ParetoBench/ParetoBench/plots")
source("custom_ecdf.R")
source("draw_paretos.R")
```

Assume the ParetoBench results are stored in ~/Documents/phd/code/ParetoBench/examples/pareto_comparison, let's set our working directory there and draw the plots.

```r
setwd("~/Documents/phd/code/ParetoBench/examples/pareto_comparison")
```

And start drawing

```r
draw_cdf(dataset="Levine13", 
         colours=c("#FF8900", "#0774A9"), 
         blank_axis=FALSE,
         save_to_disk=FALSE
        )
```

You have a CDF plot! The dataset refers to the result file for which dataset. 
If you want to draw one for all dataset, just set that to all_datasets.

You can supply custom colours (in form of hex) if you want custom colours.

Option to have blank axis, and whether to save the plot or not!

```r
draw_swarm(dataset="Levine13", 
         colours=c("#FF8900", "#0774A9"), 
         blank_axis=FALSE,
         save_to_disk=FALSE
        )
```

This is the swarm plot. The arguments are all the same as cdf plots.
